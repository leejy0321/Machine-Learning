```python
from sklearn import datasets
raw_cancer = datasets.load_breast_cancer()
```

```python
X = raw_cancer.data
y = raw_cancer.target
```

```python
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te = train_test_split(X,y,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)
```

```python
import numpy as np
import math

W_new = np.zeros(30)
W0_new = 0
a = 1e-8
epochs = 60

MSE = np.array([])

def sigmoid(output):
    z = 1/(1+math.exp(-output))
    return z

for epoch in range(epochs):

    p_preds = np.array([])
    p_pred_exps = np.array([])
    error = np.array([])
    error_x = np.zeros(30)
    p_class = np.array([])

    W = W_new
    W0 = W0_new

    for x in X_tn:

        p_pred = W0 + np.dot(W, x)
        p_preds = np.append(p_preds, p_pred)

        p_pred_exp = sigmoid(p_pred)
        p_pred_exps = np.append(p_pred_exps, p_pred_exp)

        if p_pred_exp > 0.5:
            p_class = np.append(p_class,1)
        else:
            p_class = np.append(p_class,0)

    error = p_pred_exps - y_tn
    error_x = np.dot(error, X_tn)
    
    MSE_val = (error).mean()
    MSE = np.append(MSE, MSE_val)

    W0_new = W0 - a*np.sum(error)

    for i in range(30):
        W_new[i] = W[i] - a*np.sum(error_x[i])
    
    # count = 0
    # if (epoch % 5 == 0):
    #     for i in range(len(p_class)):
    #         if ((p_class[i] - y_tn[i]) != 0):
    #             count += 1
    #     print(count, epoch)

def testset_score(X, Y):

    p_preds = np.array([])
    p_pred_exps = np.array([])
    p_class = np.array([])

    W = W_new
    W0 = W0_new

    for x in X:

        p_pred = W0 + np.dot(W, x)
        p_preds = np.append(p_preds, p_pred)

        p_pred_exp = sigmoid(p_pred)
        p_pred_exps = np.append(p_pred_exps, p_pred_exp)

        if p_pred_exp > 0.5:
            p_class = np.append(p_class,1)
        else:
            p_class = np.append(p_class,0)

    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(p_class)):
        if ((p_class[i] == 1) and (Y[i] == 1)):
            TP += 1
        elif ((p_class[i] == 1) and (Y[i] == 0)):
            FP += 1
        elif ((p_class[i] == 0) and (Y[i] == 0)):
            TN += 1
        elif ((p_class[i] == 0) and (Y[i] == 1)):
            FN += 1

    print(f'TP: {TP}, FP: {FP}, TN: {TN}, FN: {FN}')

testset_score(X_te, y_te)
```
